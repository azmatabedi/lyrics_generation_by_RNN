#tokenizer is use process of spliting the text into small sentence or spliting the apiece into set of chrachter

```python
tokenizer=Tokenizer(oov_token='OOV')
tokenizer.fit_on_texts(corpus2)
total_words=len(tokenizer.word_index)+1
```

#total  number of word after  the tokenize

```python
print("total number of words",total_words)
```

# change the words into index values beacuse it will easy to get the text form the index value

```python
reversedwordindex= {v: k for k,v in tokenizer.word_index.items()}
```

#now i am generating ngram text

```python
input_sequences=[]
for line in corpus2:
  token_list=tokenizer.texts_to_sequences([line])[0]
  for i in range(1,len(token_list)):
    n_gram_sequence=token_list[:i+1]
    input_sequences.append(n_gram_sequence)
```

```python
print('total number of generated sequence',len(input_sequences))
```

```python
input_sequences[0:4]
```

```python
max_len_sequence=max([len(x)for x in input_sequences])
```

```python
print("maximum length of a sequence",max_len_sequence)
```

#now saveing the sequence which will help me to train the model

```python
input_sequences=np.array(pad_sequences(input_sequences,maxlen=max_len_sequence,padding='pre'))
```

```python
print(input_sequences)
```

# I am seperating the X varibale and Y variable

```python
xs=input_sequences[:,:-1]
labels=input_sequences[:,-1]
```

```python
ys=tf.keras.utils.to_categorical(labels,num_classes=total_words)
```

```python
embeddingSize = 500
nodeCount = 150
```

```python
model = Sequential()
model.add(Embedding(total_words, embeddingSize, input_length=max_len_sequence-1))
model.add(Bidirectional(LSTM(units=nodeCount, return_sequences = True)))
model.add(Bidirectional(LSTM(units=nodeCount, return_sequences = True)))
model.add(LSTM(nodeCount))
model.add(Dense(total_words, activation='relu'))
model.add(Dense(total_words, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
def generate_seq(model,tokenizer,enter_text,n_pred):
  in_text,result=enter_text,enter_text
  for i in range(n_pred):
    encoded=tokenizer.texts_to_sequences([in_text])[0]
    encoded=np.array(encoded)
    yhat=np.argmax(model.predict(encoded),axis=-1)
    
    out_word=' '
    for word,index in tokenizer.word_index.items():
      if (index==yhat).any():
        out_word=word
        break
    in_text,result=out_word,result +' '+ out_word
    return result
```

# now I have the made the RNN neural network in Which i gave the embedded layer  and then add the LSTM layer layer and finaly i add the dence layer 

```python
model1 = Sequential()
model1.add(Embedding(total_words, 64, input_length = max_len_sequence -1))
model1.add(Bidirectional(LSTM(100)))
model1.add(Dense(total_words, activation='relu'))
model1.add(Dense(total_words, activation = 'softmax'))
model1.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
print(model1.summary())
```

```python
history=model1.fit(xs, ys, epochs = 80, verbose = 1)
```
